```python
import geopandas as gpd
import plotly.express as px
```

```python
# Membuka layer bangunan dari file .osm
buildings = gpd.read_file("map2.osm", layer='multipolygons')
print(buildings.head())

# Simpan ke CSV
buildings.to_csv("buildings_from_osm2.csv", index=False)
```

```python
# Process buildings from OSM
buildings_projected = buildings.to_crs(epsg=32749)  # Project to UTM for accurate area
buildings['area'] = buildings_projected.geometry.area
buildings['centroid'] = buildings.geometry.centroid
buildings['latitude'] = buildings.centroid.y
buildings['longitude'] = buildings.centroid.x

# Define variables for plotting OSM buildings
df_osm = buildings
lat_col_osm = 'latitude'
lon_col_osm = 'longitude'
kolom_osm = 'area'  # or another column, e.g., 'name' or add area
title_prefix_osm = 'OSM Buildings '

# Convert to categorical for color if needed
df_osm[kolom_osm] = df_osm[kolom_osm].astype('category')
```

```python
# Scatter Mapbox for OSM Buildings
if lat_col_osm in df_osm.columns and lon_col_osm in df_osm.columns:
    fig_osm = px.scatter_mapbox(
        df_osm,
        lat=lat_col_osm,
        lon=lon_col_osm,
        color=kolom_osm,
        color_continuous_scale="Inferno" if df_osm[kolom_osm].dtype in ['int64', 'float64'] else None,
        zoom=10,
        height=500,
        width=800,
        title=f'{title_prefix_osm}Scatter Mapbox {kolom_osm}',
        mapbox_style="open-street-map"
    )
    fig_osm.show()
```
